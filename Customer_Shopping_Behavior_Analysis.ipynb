```python
#create a column age group
labels =['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
#new purchase frequency days column textual frequency to number of days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# =========================
# RFM ANALYSIS (FULL VERSION)
# =========================

import pandas as pd
import numpy as np

# -------------------------
# 1. RECENCY SCORE (R)
# -------------------------
# Lower purchase_frequency_days = more recent

df['R_score'] = pd.cut(
    df['purchase_frequency_days'],
    bins=[0, 14, 30, 90, 365, 400],
    labels=[5, 4, 3, 2, 1]
)

# -------------------------
# 2. FREQUENCY SCORE (F)
# -------------------------
df['F_score'] = pd.qcut(
    df['previous_purchases'],
    5,
    labels=[1, 2, 3, 4, 5],
    duplicates='drop'
)

# -------------------------
# 3. MONETARY SCORE (M)
# -------------------------
df['M_score'] = pd.qcut(
    df['purchase_amount'],
    5,
    labels=[1, 2, 3, 4, 5],
    duplicates='drop'
)

# -------------------------
# 4. COMBINE RFM SCORE
# -------------------------
df['R_score'] = df['R_score'].astype(int)
df['F_score'] = df['F_score'].astype(int)
df['M_score'] = df['M_score'].astype(int)

df['RFM_score'] = (
    df['R_score'].astype(str) +
    df['F_score'].astype(str) +
    df['M_score'].astype(str)
)

# -------------------------
# 5. CUSTOMER SEGMENTATION
# -------------------------

def segment_customer(row):
    if row['R_score'] >= 4 and row['F_score'] >= 4:
        return 'Champions'
    elif row['F_score'] >= 4:
        return 'Loyal Customers'
    elif row['M_score'] >= 4:
        return 'Big Spenders'
    elif row['R_score'] <= 2:
        return 'At Risk'
    else:
        return 'Potential Loyalist'

df['customer_segment'] = df.apply(segment_customer, axis=1)

# -------------------------
# 6. SUMMARY OUTPUT
# -------------------------
segment_summary = df.groupby('customer_segment').agg(
    total_customers=('customer_id', 'count'),
    total_revenue=('purchase_amount', 'sum'),
    avg_spend=('purchase_amount', 'mean'),
).reset_index()

segment_summary['revenue_share_%'] = (
    segment_summary['total_revenue'] / segment_summary['total_revenue'].sum()
) * 100

segment_summary.sort_values(by='total_revenue', ascending=False)
```
